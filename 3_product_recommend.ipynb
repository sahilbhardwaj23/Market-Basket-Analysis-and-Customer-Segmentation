```python
import pandas as pd
import numpy as np

df=pd.read_csv('training_dataset.csv')
```

```python
def starts_with_alphabet(row):
    # Exclude the 'Invoice Number' column from the check
    filtered_row = row.drop(['Customer Id'], errors='ignore')
    return filtered_row.dropna().astype(str).str[0].str.isalpha().any()
```

```python
# Apply the function to filter out rows
df = df[df.apply(starts_with_alphabet, axis=1)]
```

```python
df.shape
```

```python
transactions = []
for i in range(0,4372):
    transactions.append([str(df.values[i,j]) for j in range(0,1885)])
```

### Implementation of the Apriori Algorithm

We imported the `apriori` function from the `apyori` module and stored the resulting output in the variable `rules`. To the `apriori` function, we passed six parameters:

1. **Transactions**: The main input is the list of transactions, which contains the products bought together.

2. **Minimum Support**: We set this to **0.002**. This value was calculated based on the assumption that a product should appear in at least 3 transactions per day. Given our data spans a week, the calculation is as follows:

   Minimum Support=3*7/7500=0.0028
  
   We rounded this to 0.002 to include relevant associations.

3. **Minimum Confidence**: We chose a minimum confidence of **0.25**, which was derived from analyzing various results to ensure that the associations are meaningful.

4. **Minimum Lift**: Set to **2.5**, indicating that the rule must be more than 2.5 times as likely to occur than if the items were independent. This helps in identifying stronger associations.

5. **Minimum Length**: Set to **2**, focusing on pairs of items. This parameter ensures that we are only calculating lift values for combinations of two items (A and B) being bought together.

6. **Maximum Length**: Also set to **2**, which means we are limiting the results to pairs of items.



```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.25, min_lift = 2.5, min_length = 2, max_length = 2)
```

```python
results=list(rules)
```

```python
# Extract relevant information from the apriori results
def extract_association_rules(results):
    rules_list = []
    for result in results:
        for ordered_stat in result.ordered_statistics:
            # Base product and associated product
            base_items = tuple(ordered_stat.items_base)
            add_items = tuple(ordered_stat.items_add)
            confidence = ordered_stat.confidence
            lift = ordered_stat.lift

            # Create a dictionary of rules for easy reference
            rules_list.append({
                'Base Item': base_items,
                'Add Item': add_items,
                'Confidence': confidence,
                'Lift': lift
            })
    return pd.DataFrame(rules_list)

# Convert rules into a DataFrame
rules_df = extract_association_rules(results)

# Show the rules

```

```python
# Function to recommend products based on a selected product
def recommend_products(product, rules_df, top_n=5):
    recommendations = rules_df[rules_df['Base Item'].apply(lambda x: product in x)]
    # Sort by highest confidence or lift
    recommendations = recommendations.sort_values(by='Confidence', ascending=False)
    
    return recommendations.head(top_n)

# Example: Recommend products based on product 'I'
product = 'ZINC T-LIGHT HOLDER STAR LARGE'
recommended_products = recommend_products(product, rules_df)

recommended_products
```
